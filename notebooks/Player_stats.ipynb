```python
columns = ['Height(sm)', 'Weight(kg)', 'Value(€)']
p[columns] = p[columns].replace('-', np.nan)
```

```python
p['Height(sm)'] = pd.to_numeric(p['Height(sm)'], errors='coerce').astype('Int64')
p['Weight(kg)'] = pd.to_numeric(p['Weight(kg)'], errors='coerce').astype('Int64')
p['Value(€)'] = pd.to_numeric(p['Value(€)'], errors='coerce').astype('Int64')
```

```python
p[['Height(sm)', 'Weight(kg)', 'Value(€)']].dtypes
```

```python
export_p = p.copy()
for column in columns:
    export_p[column] = export_p[column].astype(object).fillna('-')
```

```python
p['Position'] = p['Position'].str.strip()
```

```python
p['Position'] = p['Position'].str.replace('Centre Back', 'Center Back', regex=False)
p['Position'] = p['Position'].str.replace('Center back', 'Center Back', regex=False)
```

```python
p['Position'] = p['Position'].str.split('/').str[0]
p['Short_Position'] = p['Short_Position'].str.split('/').str[0]
p['Position'] = p['Position'].str.strip()
p['Short_Position'] = p['Short_Position'].str.strip() #Data Simplification
```

```python
p['Short_Position'] = p['Short_Position'].str.replace(r'\d+', '', regex=True)
```

```python
p['Player'] = p['Player'] \
    .str.replace('Ə', 'A', regex=False).str.replace('ə', 'a', regex=False) \
    .str.replace('Ç', 'Ch', regex=False).str.replace('ç', 'ch', regex=False) \
    .str.replace('Ş', 'Sh', regex=False).str.replace('ş', 'sh', regex=False)
```

```python
p['Player'].unique()
```

```python
p['Short_Position'] = p['Short_Position'].replace('DM', 'CDM')
```

```python
p['Start_year'] = p['Start_year'].replace('-', np.nan)
p['Start_year'] = pd.to_numeric(p['Start_year'], errors='coerce').astype('Int64')
```

```python
p['Start_year'] = p['Contract_date'].astype(str).str.extract(r'(\d{4})')[0]
```

```python
p['Start_year'].unique()
```

```python
date_columns = ['Date of birth', 'Contract_date', 'Contract_expires']
for column in date_columns:
    if p[column].dtype == 'object':
        p[column] = pd.to_datetime(p[column], errors='coerce')
    p[column] = p[column].dt.date
print(p[date_columns].head())
```

```python
Q1 = p['Weight(kg)'].quantile(0.25)
Q3 = p['Weight(kg)'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = p[ ((p['Weight(kg)'] < lower_bound) | (p['Weight(kg)'] > upper_bound)) & (~p['Weight(kg)'].between(55, 100))]
print(outliers)
print(outliers[['Player', 'Weight(kg)']])
```

```python
p.info()
```

```python
p['Height(sm)'] = p['Height(sm)'].replace('-', np.nan)
mean_height = p['Height(sm)'].mean()
p['Height(sm)'] = p['Height(sm)'].fillna(round(mean_height)).astype('Int64')
print(p['Height(sm)'].head(10))
```

```python
mean_height
```

```python
#z-score
mean = p['Weight(kg)'].mean()
std = p['Weight(kg)'].std()
p['z_score'] = (p['Weight(kg)'] - mean) / std
outliers = p[np.abs(p['z_score']) > 3]
print(f"Number of outliers found: {len(outliers)}")
print(outliers[['Player', 'Weight(kg)', 'z_score']])
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(x=p['Weight(kg)'], kde=True)
plt.axvline(x=mean + 3*std, color='red', linestyle='--', label='Upper Limit (+3)')
plt.axvline(x=mean - 3*std, color='green', linestyle='--', label='Lower Limit (-3)')
plt.legend()
plt.title('Weight Distribution and Z-score Limits')
plt.show()
```

```python
p = p.drop(outliers.index)
p = p.reset_index(drop=True)
p = p.drop(columns=['z_score'])
print(f"Number of rows in the cleaned table: {len(p)}")
```

```python
#Feature Engineering  #BMİ
p['BMI'] = p['Weight(kg)'] / ((p['Height(sm)'] / 100) ** 2)
columns = ['Goals', 'Assists', 'Minutes_played']
for column in columns:
    p[column] = pd.to_numeric(p[column], errors='coerce').fillna(0)
#Mehsuldarliq
p['Productivity'] = ((p['Goals'] + p['Assists']) / p['Minutes_played']) * 90
p['Productivity'] = p['Productivity'].fillna(0)
p['BMI'] = p['BMI'].round(2)
p['Productivity'] = p['Productivity'].round(2)
p.fillna(0, inplace=True)
print(p[['Player', 'BMI', 'Productivity']].head())
```

```python
plt.figure(figsize=(13, 7))
sns.scatterplot(
    data=p,
    x='Value(€)',
    y='Goals+Assists',
    alpha=0.6,
    s=100,
    color='green')
plt.title('Value vs Performance', fontsize=16)
plt.xlabel('Market Value (€)', fontsize=12)
plt.ylabel('Goals + Assists', fontsize=12)
plt.axvline(x=p['Value(€)'].median(), color='blue', linestyle='--', label='Average Value')
plt.axhline(y=p['Goals+Assists'].mean(), color='red', linestyle='--', label='Average Performance')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show(); #correlation
```

```python
from scipy import stats
starters = p[p['Starting eleven'] > 10]['Value(€)'].dropna()
benchers = p[p['Starting eleven'] <= 10]['Value(€)'].dropna()
#T-test
t_stat, p_value = stats.ttest_ind(starters, benchers)
print(f"P-value: {p_value:.5f}")
if p_value < 0.05:
    print("Result: Həqiqətən də Start heyət oyunçuları daha bahalıdır.")
else:
    print("Result: Ciddi bir fərq yoxdur, təsadüf ola bilər.")
```

```python
from scipy import stats
laliga_values = p[p['League'] == 'Laliga']['Value(€)'].dropna()
superlig_values = p[p['League'] == 'Super League']['Value(€)'].dropna()
#T-test
t_stat, p_value = stats.ttest_ind(laliga_values,superlig_values, equal_var=False)
print(f"Laliga Average Value: {laliga_values.mean():,.0f} €")
print(f"Super League Average Value: {superlig_values.mean():,.0f} €")
print("-" * 30)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.10f}")
print("-" * 30)
alpha = 0.05
if p_value <= alpha:
    print("Result: H₀ rədd edilir.Bu iki liqanın bazar dəyərləri arasında ciddi fərq var.")
else:
    print("Result: H₀ qəbul edilir.Aradakı fərq təsadüfi ola bilər, statistik olaraq əhəmiyyətli deyil")
```
